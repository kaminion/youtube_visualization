```python
# 영상 제목 추출 및 정제
# 데이터 조회
KRvideo = pd.read_csv('./data/KRvideos_utf8.csv', 
encoding="utf8")
```

```python
# 제목과 뷰만 가져옴

df = KRvideo[['title', 'views']]
df_sorted = df.sort_values(by="views", ascending=False).drop_duplicates(['title'], keep="first")

df_sorted
```

```python
# 제목 Array로 조회
df_sorted['title'].values
```

```python
import re
# 각각의 row에 대해서 적용
df_sorted['title_refined'] = df_sorted['title'].apply(lambda x: re.sub('[^가-힣\s]', '', x))
df_sorted
```

```python
df_sorted = df_sorted[df_sorted['title_refined'].apply(lambda x: re.sub('[^가-힣]', '', x)) != '']
df_sorted
```

```python
# KoNLPy, soynlp 한글 데이터 가공라이브러리
# KoNLPy 미등록 단어 문제로 새로운 단어를 인식불가(ex. 너무너무너무)
# soynlp - 학습기반 

# WordExtractor
# 통계 기반 단어 추출기 (문서집합에서 통계 기반 자주 등장하는 단어열 추출)

# Tokenizer
# Word Extractor의 결과를 이용하여 단어의 경계를 따라
# 문장을 단어열로 분해
# LTokenizer - 띄어쓰기 잘되어있는경우, MaxScoreTokenizer - 띄어쓰기 잘 안되어있는경우
```

```python
# 짜장면/명사(L) + 을/조사(R)
# 먹/동사(L) + 었어/어미(R)
```

```python
from soynlp.word import WordExtractor

word_extractor = WordExtractor(min_frequency=100,
   min_cohesion_forward=0.05,
   min_right_branching_entropy=0.0
)
word_extractor.train(df_sorted['title_refined'].values)
words = word_extractor.extract()
words
```

```python
# word extractor로 단어 점수를 계산하였다면 진행

from soynlp.word import WordExtractor
from soynlp.utils import DoublespaceLineCorpus
from soynlp.tokenizer import LTokenizer

cohesion_score = {word:score.cohesion_forward for word, score in words.items()}
tokenizer = LTokenizer(scores=cohesion_score)

tokenizer
```

```python
df_sorted['tokenized'] = df_sorted['title_refined'].apply(lambda x: tokenizer.tokenize(x, remove_r=True)) # L과 R 분리하여 R은 삭제
df_sorted
```

```python
# 인기영상 키워드 분석
df_top = df_sorted[df_sorted['views'] > 500000].reset_index()
df_top
```

```python
# tokenized 단어 한 배열에 담기
words = []
for i in df_top['tokenized'].values:
    for k in i:
        words.append(k) # 각각의 단어 list append
print(words)
```

```python
# Counter로 words 빈도수 파악, 파이썬 내장 라이브러리
from collections import Counter
count = Counter(words)
words_dict = dict(count)
words_dict
```

```python
# 워드 크라우드란 자주 쓰는 단어를 크게 보여주는..
# 단어의 빈도수에 따라 그림
# stop word : 스잘데기없는 단어

!pip install WordCloud
```
